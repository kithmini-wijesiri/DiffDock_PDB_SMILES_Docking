```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

**Scatter Plots**

1. creates a scatter plot showing the relationship between the total bill and the tip amount in the tips dataset

```python
#scatter -> axis level function
sns.color_palette("pastel")
sns.scatterplot(x='total_bill', y='tip', data=tips)
```

2. creates a scatter plot showing the relationship between the total bill and the tip amount in the tips dataset. Each point on the plot represents a data point from the dataset, with the x-coordinate representing the total bill and the y-coordinate representing the tip amount. This allows you to visually explore how the tip amount varies with the total bill.

```python
#scatter -> figure level function
sns.relplot(x='total_bill', y='tip', data=tips, kind='scatter')
```

3. This code sets up a customized matplotlib figure, adjusts the Seaborn plotting style, and creates a scatter plot with specific visual attributes based on the provided dataset (tips). The scatter plot visualizes the relationship between the total bill and tip amount, with additional information encoded through color, marker style, and marker size.

```python
plt.figure(figsize=(10,6))
sns.set_theme(style="darkgrid", palette="Set2", font_scale=1.2)
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='sex', style='time', size='size')
plt.show()
```

4. creates a scatter plot showing the relationship between the total bill and the tip amount in the tips dataset. The points are colored based on the gender ('sex' column) of the customers, and the marker style is based on the time ('time' column) of the meal (lunch or dinner).

```python
sns.relplot(x='total_bill', y='tip', data=tips, kind='scatter', hue='sex', style='time')
```

**Line Plots**

```python
df = px.data.gapminder()
df.head()
```

```python
df1 = df[df['country'] == 'India']
df1
```

line plot using Seaborn's lineplot function to visualize life expectancy (lifeExp) over the years (year)

```python
#axis level function
plt.figure(figsize=(10,6))
sns.lineplot(x='year', y='lifeExp', data=df1)
plt.show()
```

```python
#figure level
sns.relplot(x='year', y='lifeExp', data=df1, kind='line')
```

```python
df2 = df[df['country'].isin(['India', 'Pakistan', 'China'])]
df2.head()
```

```python
sns.relplot(x='year', y='lifeExp', data=df2, kind='line', hue='country')
```

```python
sns.lineplot(x='year', y='lifeExp', data=df2, hue='country')
```

```python
df3 = df[df['country'].isin(['India', 'Brazil', 'Germany'])]
df3.head()
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(x='year', y='lifeExp', data=df3, hue='country', style='continent')
plt.show()
```
